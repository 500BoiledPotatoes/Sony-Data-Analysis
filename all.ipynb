```python

import os
import pandas as pd
from plotly.offline import iplot, init_notebook_mode
import cufflinks as cf
cf.go_offline(connected=True)
init_notebook_mode(connected=True)
path = 'files'
path_read_encoder = []
path_read_decoder = []

def read(file_path):
    temp_list = os.listdir(file_path)
    for temp_list_each in temp_list:
        if os.path.isfile(file_path + '/' + temp_list_each):
            temp_path = file_path + '/' + temp_list_each
            if os.path.splitext(temp_path)[-1] == '.csv':
                if temp_path.split('/')[1][0] == 'd':
                    path_read_decoder.append(temp_path)
                elif temp_path.split('/')[1][0] == 'e':
                    path_read_encoder.append(temp_path)
            else:
                continue
        else:
            read(file_path + '/' + temp_list_each)


read(path)
def decoder_analysis():
    for file in path_read_decoder:
        df = pd.read_csv(file)
        print(filter_by_DropRate_notzero(df))
        rtt_calculate(df)
        df[['DateTime', 'RTT', 'RecvBufs']].iplot(
        y='RTT',
        x='DateTime',
        mode = 'lines',
        secondary_y = 'RecvBufs',
        opacity=0.8,
        symbol = 1,
        xTitle = 'Date',
        yTitle = 'RTT',
        secondary_y_title = 'RecvBufs',
        title = file.split('/')[1]

    )

def decoder_single_analysis(file):
    df = pd.read_csv(file)
    print(filter_by_DropRate_notzero(df))
    rtt_calculate(df)
    df[['DateTime', 'RTT', 'RecvBufs']].iplot(
        y='RTT',
        x='DateTime',
        mode = 'lines',
        secondary_y = 'RecvBufs',
        opacity=0.8,
        symbol = 1,
        xTitle = 'Date',
        yTitle = 'RTT',
        secondary_y_title = 'RecvBufs',
        title = '1'
        )

def encoder_single_analysis(file):
    df = pd.read_csv(file)
    print(filter_by_DropRate_notzero(df))
    rtt_calculate(df)
    df[['DateTime', 'RTT', 'SendBufs']].iplot(
        y='RTT',
        x='DateTime',
        mode = 'lines',
        secondary_y = 'SendBufs',
        opacity=0.8,
        symbol = 1,
        xTitle = 'Date',
        yTitle = 'RTT',
        secondary_y_title = 'SendBufs',
        title = '1'
        )

def encoder_analysis():
    for file in path_read_encoder:
        df = pd.read_csv(file)
        print(filter_by_DropRate_notzero(df))
        rtt_calculate(df)
        df[['DateTime', 'RTT', 'SendBufs']].iplot(
            x='DateTime',
            y='RTT',
            mode='lines',
            secondary_y='SendBufs',
            opacity=0.8,
            symbol=1,
            xTitle='Date',
            yTitle='RTT',
            secondary_y_title='SendBufs',
            title=file.split('/')[1]
        )

def decoder_single_analysis_range(file, num):
    df = pd.read_csv(file)
    print(filter_by_DropRate_notzero(df))
    rtt_calculate(df)
    find_rtt(df, num)
    df[['DateTime', 'RTT', 'RecvBufs']].iplot(
        y='RTT',
        x='DateTime',
        mode = 'lines',
        secondary_y = 'RecvBufs',
        opacity=0.8,
        symbol = 1,
        xTitle = 'Date',
        yTitle = 'RTT',
        secondary_y_title = 'RecvBufs',
        title = '1'
        )

def encoder_single_analysis_range(file, num):
    df = pd.read_csv(file)
    print(filter_by_DropRate_notzero(df))
    rtt_calculate(df)
    find_rtt(df, num)
    # bw_calculate(df)
    df[['DateTime', 'RTT', 'SendBufs']].iplot(
        y='RTT',
        x='DateTime',
        mode = 'lines',
        secondary_y = 'SendBufs',
        opacity=0.8,
        symbol = 1,
        xTitle = 'Date',
        yTitle = 'RTT',
        secondary_y_title = 'SendBufs',
        title = '1'
        )
def merge(file1, file2):
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    # pd.concat([df1, df2],ignore_index=True,join="outer",keys=["df1","df2"]).sort_values(by='DateTime')
    # # print(pd.concat([df1, df2],ignore_index=True,join="outer",keys=["df1","df2"]).sort_values(by='DateTime'))
    # return pd.concat([df1, df2],ignore_index=True,join="outer",keys=["df1","df2"]).sort_values(by='DateTime').drop_duplicates("DateTime")
    merge_df = pd.concat([df1, df2],ignore_index=True)
    merge_df_drop = merge_df.drop_duplicates(subset="DateTime",keep="last").sort_values(by='DateTime')
    # print(file1.split('/')[1].split('.')[0])
    print('files_merged/merged_'+ file1.split('/')[1].split('.')[0] + '_' + file2.split('/')[1].split('.')[0] + ".csv")
    merge_df_drop.to_csv('files_merged/merged_'+ file1.split('/')[1].split('.')[0] + '_' + file2.split('/')[1].split('.')[0] + ".csv", index=False)
    if file1.split('/')[1][0] == 'd':
        decoder_single_analysis('files_merged/merged_'+ file1.split('/')[1].split('.')[0] + '_' + file2.split('/')[1].split('.')[0] + ".csv")
    elif file1.split('/')[1][0] == 'e':
        encoder_single_analysis('files_merged/merged_'+ file1.split('/')[1].split('.')[0] + '_' + file2.split('/')[1].split('.')[0] + ".csv")
    return merge_df_drop

def filter_by_time(file, start_time, end_time, num):
    df = pd.read_csv(file)
    filtered_df = df[(df["DateTime"] >= start_time) & (df['DateTime'] <= end_time)]
    filtered_df.to_csv('files_filter_by_time/filtered_'+ file.split('/')[1].split('.')[0]+ ".csv", index=False)
    print('files_filter_by_time/filtered_'+ file.split('/')[1].split('.')[0]+ ".csv")
    if file.split('/')[1][0] == 'd':
        decoder_single_analysis_range('files_filter_by_time/filtered_'+ file.split('/')[1].split('.')[0]+ ".csv", num)
    elif file.split('/')[1][0] == 'e':
        encoder_single_analysis_range('files_filter_by_time/filtered_'+ file.split('/')[1].split('.')[0]+ ".csv", num)
    return filtered_df

def filter_by_DropRate_notzero(df):
    filter_df = df[df['DropRate'] != 0]
    print("DropRate不为0数据: ")
    return filter_df.to_string(index=False, line_width=10000)

def rtt_calculate(df):
    max_value = df['RTT'].max()
    min_value = df['RTT'].min()
    mean_value = df['RTT'].mean()
    date_time_max = df.loc[df['RTT'] == max_value, 'DateTime']
    date_time_min = df.loc[df['RTT'] == min_value, 'DateTime']
    # datetime_values = date_time_max.tolist()
    # datetime_values = pd.Series(date_time_max)
    # datetime_values_min = date_time_min.tolist()
    # datetime_values_min = pd.Series(date_time_min)
    print('RTT最大值:')
    filtered_data_max = df.loc[df['RTT'] == max_value, ['DateTime', 'RTT']]
    print(filtered_data_max.to_string(index=False, line_width=10000))
    print('RTT最小值:')
    filtered_data_min = df.loc[df['RTT'] == min_value, ['DateTime', 'RTT']]
    print(filtered_data_min.to_string(index=False, line_width=10000))
    # print('时间：', datetime_values, 'RTT最大值:', max_value)
    # print('时间：', datetime_values_min, 'RTT最小值:', min_value)
    print('RTT平均值:', mean_value)

def bw_calculate(df):
    max_value = df['MaxBw'].max()
    min_value = df['MaxBw'].min()
    mean_value = df['MaxBw'].mean()
    date_time_max = df.loc[df['MaxBw'] == max_value, 'DateTime']
    date_time_min = df.loc[df['MaxBw'] == min_value, 'DateTime']
    # datetime_values = date_time_max.tolist()
    # datetime_values = pd.Series(date_time_max)
    # datetime_values_min = date_time_min.tolist()
    # datetime_values_min = pd.Series(date_time_min)
    print('MaxBw最大值:')
    filtered_data_max = df.loc[df['MaxBw'] == max_value, ['DateTime', 'RTT']]
    print(filtered_data_max.to_string(index=False, line_width=10000))
    print('MaxBw最小值:')
    filtered_data_min = df.loc[df['MaxBw'] == min_value, ['DateTime', 'RTT']]
    print(filtered_data_min.to_string(index=False, line_width=10000))
    # print('时间：', datetime_values, 'RTT最大值:', max_value)
    # print('时间：', datetime_values_min, 'RTT最小值:', min_value)
    print('MaxBw平均值:', mean_value)

def find_rtt(df, rtt):
    filtered_data = df[df['RTT'] > rtt]
    print("RTT范围内数据: ")
    print(filtered_data.to_string(index=False, line_width=10000))
```
