```python

import os
import pandas as pd
from plotly.offline import iplot, init_notebook_mode
import cufflinks as cf
cf.go_offline(connected=True)
init_notebook_mode(connected=True)
path = 'files'
path_read_encoder = []
path_read_decoder = []

def read(file_path):
    temp_list = os.listdir(file_path)
    for temp_list_each in temp_list:
        if os.path.isfile(file_path + '/' + temp_list_each):
            temp_path = file_path + '/' + temp_list_each
            if os.path.splitext(temp_path)[-1] == '.csv':
                if temp_path.split('/')[1][0] == 'd':
                    path_read_decoder.append(temp_path)
                elif temp_path.split('/')[1][0] == 'e':
                    path_read_encoder.append(temp_path)
            else:
                continue
        else:
            read(file_path + '/' + temp_list_each)


read(path)
def decoder_analysis():
    for file in path_read_decoder:
        df = pd.read_csv(file)
        print(filter_by_DropRate_notzero(df))
        df[['DateTime', 'RTT', 'RecvBufs']].iplot(
        y='RTT',
        x='DateTime',
        mode = 'lines',
        secondary_y = 'RecvBufs',
        opacity=0.8,
        symbol = 1,
        xTitle = 'Date',
        yTitle = 'RTT',
        secondary_y_title = 'RecvBufs',
        title = file.split('/')[1]

    )

def decoder_single_analysis(file):
    df = pd.read_csv(file)
    print(filter_by_DropRate_notzero(df))
    df[['DateTime', 'RTT', 'RecvBufs']].iplot(
        y='RTT',
        x='DateTime',
        mode = 'lines',
        secondary_y = 'RecvBufs',
        opacity=0.8,
        symbol = 1,
        xTitle = 'Date',
        yTitle = 'RTT',
        secondary_y_title = 'RecvBufs',
        title = '1'
        )

def encoder_single_analysis(file):
    df = pd.read_csv(file)
    print(filter_by_DropRate_notzero(df))
    df[['DateTime', 'RTT', 'SendBufs']].iplot(
        y='RTT',
        x='DateTime',
        mode = 'lines',
        secondary_y = 'RecvBufs',
        opacity=0.8,
        symbol = 1,
        xTitle = 'Date',
        yTitle = 'RTT',
        secondary_y_title = 'RecvBufs',
        title = '1'
        )

def encoder_analysis():
    for file in path_read_encoder:
        df = pd.read_csv(file)
        filter_by_DropRate_notzero(df)
        print(filter_by_DropRate_notzero(df))
        df[['DateTime', 'RTT', 'SendBufs']].iplot(
            x='DateTime',
            y='RTT',
            mode='lines',
            secondary_y='SendBufs',
            opacity=0.8,
            symbol=1,
            xTitle='Date',
            yTitle='RTT',
            secondary_y_title='SendBufs',
            title=file.split('/')[1]
        )

def merge(file1, file2):
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    # pd.concat([df1, df2],ignore_index=True,join="outer",keys=["df1","df2"]).sort_values(by='DateTime')
    # # print(pd.concat([df1, df2],ignore_index=True,join="outer",keys=["df1","df2"]).sort_values(by='DateTime'))
    print(pd.concat([df1, df2],ignore_index=True))

    # return pd.concat([df1, df2],ignore_index=True,join="outer",keys=["df1","df2"]).sort_values(by='DateTime').drop_duplicates("DateTime")
    merge_df = pd.concat([df1, df2],ignore_index=True)
    merge_df_drop = merge_df.drop_duplicates(subset="DateTime",keep="last").sort_values(by='DateTime')
    # print(file1.split('/')[1].split('.')[0])
    print('files/merged_'+ file1.split('/')[1].split('.')[0] + '_' + file2.split('/')[1].split('.')[0] + ".csv")
    merge_df_drop.to_csv('files/merged_'+ file1.split('/')[1].split('.')[0] + '_' + file2.split('/')[1].split('.')[0] + ".csv")
    if file1.split('/')[1][0] == 'd':
        decoder_single_analysis('files/merged_'+ file1.split('/')[1].split('.')[0] + '_' + file2.split('/')[1].split('.')[0] + ".csv")
    elif file1.split('/')[1][0] == 'e':
        encoder_single_analysis('files/merged_'+ file1.split('/')[1].split('.')[0] + '_' + file2.split('/')[1].split('.')[0] + ".csv")
    return merge_df_drop

def filter_by_time(file, start_time, end_time):
    df = pd.read_csv(file)
    # start_time = datetime.datetime.strptime(start_time,"%Y-%m-%dT%H:%M:%S%z")
    # print(start_time)
    # end_time = datetime.datetime.strptime(end_time,"%Y-%m-%dT%H:%M:%S%z")
    filtered_df = df[(df["DateTime"] >= start_time) & (df['DateTime'] <= end_time)]
    if file.split('/')[1][0] == 'd':
        decoder_single_analysis(file)
    elif file.split('/')[1][0] == 'e':
        encoder_single_analysis(file)
    filtered_df.to_csv('files_filter_by_time/filtered_'+ file.split('/')[1].split('.')[0]+ ".csv")
    print('files_filter_by_time/filtered_'+ file.split('/')[1].split('.')[0]+ ".csv")
    return filtered_df
# filter_by_time('./files/decoder-1080p2-2023-06-08-10h10m54s.csv', '2023-06-07T17:02:15+0800', "2023-06-07T17:02:55+0800")

def filter_by_DropRate_notzero(df):
    filter_df = df[df['DropRate'] != 0]
    return filter_df

# merge(path_read_decoder[3],path_read_decoder[2])
# decoder_analysis()
```
